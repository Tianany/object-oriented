```python
# QUESTION 2
import sys
from heapq import heappop, heappush # implimentation for heapsort algorithm
import networkx as nx #module used for graph visualisation
import matplotlib.pyplot as plt # graph visualisation
import tkinter as tk #module for creating GUI
from tkinter import messagebox  # module for creating pop-up messages

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v, w):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append((v, w))
        self.graph[v].append((u, w))


def dijkstra(graph, src, dest):
    """
    This function implements Dijkstra's algorithm to find the shortest path
    between two cities in a graph.

    :param graph: adjacency list representing the graph
    :param src: source city
    :param dest: destination city
    :return: shortest path between src and dest
    """
    n = len(graph.graph)  # Number of cities
    # Initialize distance array
    dist = {node: sys.maxsize for node in graph.graph}
    dist[src] = 0

    # Initialize priority queue
    pq = [(0, src)]

    # Initialize previous node array
    prev = {node: None for node in graph.graph}

    while pq:
        # Get the city with minimum distance from the source city
        d, u = heappop(pq)

        # Skip if the city has already been visited
        if d > dist[u]:
            continue

        # Update distance values of adjacent cities
        for v, w in graph.graph[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                prev[v] = u
                heappush(pq, (dist[v], v))

    # Get the shortest path by backtracking from the destination city
    path = []
    current_city = dest
    while current_city is not None:
        path.append(current_city)
        current_city = prev[current_city]
    path.reverse()
    
    # Get the total distance of the shortest path
    total_distance = dist[dest]

    return path, total_distance

    #return path

# Example usage:
graph = Graph()
graph.add_edge('A', 'B',336)
graph.add_edge('B', 'C',448)
graph.add_edge('A', 'C', 213)
graph.add_edge('A', 'D', 223)
graph.add_edge('B', 'D', 151)
graph.add_edge('E', 'F', 67)
graph.add_edge('E', 'G', 216)
graph.add_edge('F', 'G', 218)
graph.add_edge('D', 'F', 124)
graph.add_edge('C', 'D', 304)




# QUESTION 3
#import tkinter as tk

def on_find_path():
    src_city = src_var.get()
    dest_city = dest_var.get()
    print(f"Selected source city: {src_city}")
    print(f"Selected destination city: {dest_city}")
    path,total_distance = dijkstra(graph, src_city, dest_city)
    #path, total_distance = dijkstra(graph, src_city, dest_city)
    path_str = ' -> '.join(path)
    messagebox.showinfo("Shortest Path", f"The shortest path is:\n{path_str}\n\nTotal distance: {total_distance}")

    
    print(f"Shortest path: {path}")

# Create the root window
root = tk.Tk()
root.title("Shortest Path Finder")

# Set the font for all widgets
font = ("Helvetica", 16)
root.option_add("*Font", font)

# Create the source city label and dropdown
src_label = tk.Label(root, text="Select source city:")
src_label.pack(pady=10)


src_var = tk.StringVar(root)
src_var.set("select source city")  # Set default value

src_dropdown = tk.OptionMenu(root, src_var, "A", "B", "C", "D", "E", "F", "G")
src_dropdown.pack(pady=10)

dest_entry = tk.Entry(root)
dest_entry.pack()

# Create the destination city label and dropdown

dest_label = tk.Label(root, text="Select destination city:")
dest_label.pack(pady=10)


dest_var = tk.StringVar(root)
dest_var.set("select destination city")  # Set default value



dest_dropdown = tk.OptionMenu(root, dest_var, "A", "B", "C", "D", "E", "F", "G")
dest_dropdown.pack(pady=10)

dest_entry = tk.Entry(root)
dest_entry.pack()

# Create the find path button
find_path_button = tk.Button(root, text="Find Shortest Path", command=on_find_path)
find_path_button.pack(pady=10)

# Run the main loop
root.mainloop()

```
